```
import pandas as pd

# Supondo que 'dados' seja o seu DataFrame e 'SG_UF_RESIDENCIA' seja a coluna com as siglas dos estados
# Crie o dicionário de mapeamento
regioes = {
    'PA': 'Norte', 'RO': 'Norte', 'MT': 'Centro_Oeste', 'AC': 'Norte', 'AM': 'Norte', 'RR': 'Norte',
    'MA': 'Nordeste', 'PR': 'Sul', 'BA': 'Nordeste', 'PI': 'Nordeste', 'CE': 'Nordeste', 'PE': 'Nordeste',
    'AP': 'Norte', 'TO': 'Norte', 'SC': 'Sul', 'GO': 'Centro_Oeste', 'MG': 'Sudeste', 'SP': 'Sudeste',
    'SE': 'Nordeste', 'RJ': 'Sudeste', 'PB': 'Nordeste', 'AL': 'Nordeste', 'RN': 'Nordeste', 'ES': 'Sudeste',
    'DF': 'Centro_Oeste', 'RS': 'Sul', 'MS': 'Centro_Oeste'
}

# Aplicar o mapeamento para criar uma nova coluna 'Regiao'
dados['Regiao'] = dados['SG_UF_RESIDENCIA'].map(regioes)

# Exibir as primeiras linhas do DataFrame para verificar a nova coluna
dados.head()
```

```
dados['Regiao'].value_counts().plot.bar()
```

```
import matplotlib.pyplot as plt
import pandas as pd

# Suponha que 'dados' seja o seu DataFrame
# Contagem de valores por região
contagem_regioes = dados['Regiao'].value_counts()

# Criar o gráfico de barras
fig, ax = plt.subplots()
barras = contagem_regioes.plot.bar(color=[ 'green' if x in ['Sudeste', 'Nordeste'] else 'red' for x in contagem_regioes.index], ax=ax)

# Adicionar rótulos em cima das barras
for barra in barras.containers[0]:
    ax.annotate(f'{int(barra.get_height())}',  # Texto do rótulo
                (barra.get_x() + barra.get_width() / 2, barra.get_height()),  # Posição do rótulo
                ha='center', va='bottom')  # Alinhamento horizontal e ajuste vertical

# Configurações adicionais
ax.set_xlabel('Região')  # Rótulo do eixo X
ax.set_ylabel('Quantidade de Alunos')  # Rótulo do eixo Y
ax.set_title('Contagem de Alunos por Região')  # Título do gráfico
plt.xticks(rotation=0)  # Rotação dos rótulos do eixo X
plt.show()
```

```
#Plotar a quantidade de alunos por idade (NU_IDADE)
#Plote de 13 até 40 anos
#query = consulta ('filtros')
idade_13_40_anos_treineiros = dados.query('NU_IDADE >= 13 and NU_IDADE <= 40 and IN_TREINEIRO == 1')['NU_IDADE']
idade_13_40_anos_treineiros.value_counts().sort_index().plot.bar()
```

```
#Plotar as notas de redação dos homens e das mulheres (NU_NOTA_REDACAO)
redacao_das_alunas = dados.query('TP_SEXO == "F"')['NU_NOTA_REDACAO']
redacao_das_alunas.hist(bins=40)

redacao_dos_alunos = dados.query('TP_SEXO == "M"')['NU_NOTA_REDACAO']
redacao_dos_alunos.hist(bins=40)

#coloque as legendas no gráfico
```

```
import matplotlib.pyplot as plt

# Supondo que 'dados' é o seu DataFrame e já contém as colunas necessárias
notas_redacao_mulheres = dados.query('TP_SEXO == "F"')['NU_NOTA_REDACAO']
notas_redacao_homens = dados.query('TP_SEXO == "M"')['NU_NOTA_REDACAO']

# Criar o histograma
plt.figure(figsize=(10, 6))  # Ajusta o tamanho da figura
plt.hist(notas_redacao_mulheres, bins=40, alpha=0.5, label='Mulheres')  # Histograma para mulheres
plt.hist(notas_redacao_homens, bins=40, alpha=0.5, label='Homens')  # Histograma para homens

# Adicionando títulos e rótulos
plt.title('Distribuição das Notas de Redação por Sexo')
plt.xlabel('Nota de Redação')
plt.ylabel('Número de Candidatos')
plt.legend()  # Mostra a legenda

# Mostrar o gráfico
plt.show()
```
