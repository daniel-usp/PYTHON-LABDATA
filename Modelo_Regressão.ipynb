```python
import plotly.express as px
import statsmodels.api as sm
import numpy as np
import pandas as pd

# Carregando o dataset
df = pd.read_excel("exemplo.xlsx")

# Definindo as variáveis X e Y para o modelo
X = df['Distância (KM)']
Y = df['Preço do Produto']

# Adicionando constante para o modelo de regressão linear
X_with_constant = sm.add_constant(X)

# Criando o modelo de regressão linear
model = sm.OLS(Y, X_with_constant)
results = model.fit()

# Gerando a reta de regressão
df['predicted'] = results.predict(X_with_constant)

# Criando o gráfico de dispersão com a reta de regressão
fig = px.scatter(df, x='Distância (KM)', y='Preço do Produto', trendline='ols', title="Preço vs Distância com Reta de Regressão")
fig.update_traces(marker=dict(size=8))
fig.update_layout(xaxis_title='Distância (KM)', yaxis_title='Preço do Produto')

# Exibir o gráfico
fig.show()

# Exibir o relatório da análise do modelo
summary = results.summary()
print(summary)
```
